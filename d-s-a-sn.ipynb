```python
import pandas as pd
```

```python
df = pd.read_csv("survey_results_public.csv", low_memory=False)
df1 = pd.read_csv("survey_results_schema.csv")
```

```python
#Завдання 1. Підрахунок загальної кількості респондентів
total_respondents = df.shape[0]
total_respondents
```

```python
#Завдання 2. Аналіз повноти відповідей респондентів
```

```python
# Список назв питань зі схеми
schema_questions = set(df1['qname'])
len(schema_questions)
```

```python
#список колонок з основного датасету
data_columns = set(df.columns)
len(data_columns)
```

```python
#Інтерсекція множин (обовʼязково за умовою)
questions_in_data = schema_questions.intersection(data_columns)
len(questions_in_data)
```

```python
# DataFrame тільки з питаннями опитування
df_questions = df[list(questions_in_data)]
df_questions.head()
```

```python
#респонденти без пропущених значень
complete_responses = df_questions.dropna()
```

```python
# Кількість респондентів, які відповіли на всі запитання
complete_responses_count = complete_responses.shape[0]
complete_responses_count
```

```python
#Завдання 3. Статистичний аналіз досвіду респондентів
```

```python
# Обчислення мір центральної тенденції для досвіду роботи
workexp_mean = df['WorkExp'].mean()
workexp_median = df['WorkExp'].median()
workexp_mode = df['WorkExp'].mode()
```

```python
# Зведена таблиця з результатами
workexp_stats = pd.DataFrame({
    'Metric': ['Mean', 'Median', 'Mode'],
    'WorkExp': [workexp_mean, workexp_median, workexp_mode.iloc[0]]
})

workexp_stats
```

```python
#Завдання 5. Визначення популярності Python
```

```python
# Загальна кількість респондентів
total_respondents = df.shape[0]

# Респонденти, які використовують Python
python_users = df[df['LanguageHaveWorkedWith'].str.contains('Python', na=False)]

# Кількість таких респондентів
python_count = python_users.shape[0]

# Відсоток респондентів, які програмують на Python
python_percentage = (python_count / total_respondents) * 100

# Виведення результату
print(f"Відсоток респондентів, які програмують на Python: {python_percentage:.1f}%")
```

```python
#Завдання 6. Аналіз шляхів навчання програмуванню
```

```python
# Респонденти, які навчалися програмувати через онлайн курси
online_course_learners = df[
    df['LearnCode'].str.contains('Online Courses', na=False)
]

# Кількість таких респондентів
online_course_count = online_course_learners.shape[0]

# Виведення результату
online_course_count
```

```python
#Завдання 7. Географічний аналіз компенсації Python-розробників

```

```python
# 1. Фільтрація Python-розробників
python_devs = df[
    df['LanguageHaveWorkedWith'].str.contains('Python', na=False)
]

# 2. Вибір необхідних колонок та видалення пропущених значень компенсації
python_devs_comp = python_devs[['Country', 'ConvertedCompYearly']].dropna()

# 3. Групування по країнах та обчислення середньої і медіанної компенсації
comp_by_country = (
    python_devs_comp
    .groupby('Country')['ConvertedCompYearly']
    .agg(
        mean_compensation='mean',
        median_compensation='median'
    )
    .reset_index()
)

# 4. Перегляд результату
comp_by_country
```

```python
#Завдання 8. Аналіз освіти найбільш оплачуваних спеціалістів
```

```python
# Вибір необхідних колонок та видалення пропущених значень компенсації
top_paid = (
    df[['EdLevel', 'ConvertedCompYearly']]
    .dropna(subset=['ConvertedCompYearly'])
    .sort_values(by='ConvertedCompYearly', ascending=False)
    .head(5)
)

# Виведення рівнів освіти топ-5 найбільш оплачуваних респондентів
top_paid[['EdLevel']]
```

```python
#Завдання 9. Аналіз популярності Python по віковим категоріям 

```

```python
# Фільтруємо лише респондентів з вказаним віком
df_age = df.dropna(subset=['Age'])

# Функція для перевірки Python у списку мов
def is_python(user_langs):
    return 'Python' in str(user_langs)

# Додаємо колонку Python (True/False)
df_age['Python'] = df_age['LanguageHaveWorkedWith'].apply(is_python)

# Групування по віковим категоріям та обчислення відсотка Python-розробників
python_by_age = (
    df_age.groupby('Age')['Python']
    .mean()  # середнє True/False дає частку
    .reset_index()
)

# Переводимо у відсотки
python_by_age['PythonPercentage'] = (python_by_age['Python'] * 100).round(2)

# Виведення результату
python_by_age[['Age', 'PythonPercentage']].sort_values('Age')
```

```python
#Завдання 10. Аналіз індустрій серед високооплачуваних віддалених працівників
```

```python
# 1. Розрахунок 75-го перцентиля компенсації
comp_75 = df['ConvertedCompYearly'].quantile(0.75)

# 2. Фільтрація високооплачуваних віддалених працівників
high_paid_remote = df[
    (df['ConvertedCompYearly'] >= comp_75) &
    (df['RemoteWork'].str.contains('Remote', na=False))
]

# 3. Підрахунок найпопулярніших індустрій
industry_counts = (
    high_paid_remote['Industry']
    .value_counts()
    .reset_index()
    .rename(columns={'index':'Industry', 'Industry':'Count'})
)

# 4. Виведення результату
industry_counts
```
